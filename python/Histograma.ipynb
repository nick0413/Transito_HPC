```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("rol.csv",delimiter = " ", header = 0)

df.head()
```

```python
# Create a histogram of the data using Matplotlib
hist, bins, _=plt.hist(df["rol"], bins=3, edgecolor="black")
plt.title("Histograma droles")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.savefig("Histograma-roles.pdf")
plt.show()
print(hist, bins)
print(200*0.91)
print(200*0.04)
print(200*0.05)
```

```python
plt.hist(df["prob"], bins=12)
plt.title("Histogram probabilidades")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.savefig("Histograma-dGaussiana.pdf")
```
